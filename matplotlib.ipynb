```python
from matplotlib import pyplot as plt
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y)
plt.title('info')
plt.ylabel('y axix')
plt.xlabel('x axis')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x=[5,8,10]
y=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.plot(x,y,'g',label='line one',linewidth=5)
plt.plot(x2,y2,'c',label='line two',linewidth=5)
plt.title('epic info')
plt.ylabel('y axis')
plt.xlabel('x axix')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
#bar chart
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],label='audi',width=.5,color='g')
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],label='BMW',width=.5,color='r')
plt.legend()
plt.xlabel('Days')
plt.ylabel('Distance(KMS)')
plt.title('cars')
plt.grid(True,color='b')
plt.show()
```
